```python
def categorize_lexeme(lexeme):
    keywords = ["public", "static", "void", "int"]
    punctuation = ['(', ')', '{', '}', ';', ',']
    operators = ['+', '=', '==']

    if lexeme in keywords:
        return "<Keyword, {}>".format(lexeme)
    elif lexeme in punctuation:
        return "<Punctuation, {}>".format(lexeme)
    elif lexeme in operators:
        return "<Operator, {}>".format(lexeme)
    else:
        return "<Identifier, {}>".format(lexeme)

def tokenize_java_code(code):
    tokens = code.replace('(', ' ( ').replace(')', ' ) ').replace('{', ' { ').replace('}', ' } ').replace(';', ' ; ').replace(',', ' , ').split()

    categorized_tokens = []
    for token in tokens:
        categorized_token = categorize_lexeme(token)
        categorized_tokens.append(categorized_token)

    return categorized_tokens

java_code = """
Public static void main(){
int a , b ;
c = a + b;
}
"""

print("Token Separation:")
for token in tokenize_java_code(java_code):
    print(token)
```

```python
#Number of vehicles sold to retail customers
# Function to calculate the retail sales information for a range of dates
from datetime import datetime, timedelta
def vehicles_sold_per_day(start_day):
    return 1 + 2 * (start_day - 1)
def calculate_retail_sales(start_date, end_date):
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")
    total_retail_sales = 0
    retail_sales_info = []
    num_days = (end_date - start_date).days + 1
    day_counter = 0
    for day in range(num_days):
        current_date = start_date + timedelta(days=day)
        vehicles_sold_today = vehicles_sold_per_day(day_counter % 5 + 1)
        if (day_counter % 5) == 4:
            day_counter += 1
            continue
        total_retail_sales += vehicles_sold_today
        retail_sales_info.append({
            "Date": current_date.strftime("%Y-%m-%d"),
            "Vehicles Sold": vehicles_sold_today
        })
        day_counter += 1
    return retail_sales_info, total_retail_sales
retail_sales_info, total_retail_sales = calculate_retail_sales("2023-04-01", "2023-09-30")
retail_sales_info, total_retail_sales
```

```python
# Function to calculate the corporate sales information for a range of dates
def calculate_corporate_sales(start_date, end_date):
    # Parse the start and end date
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")
    
    # Initialize the results
    total_corporate_sales = 0
    corporate_sales_info = []

    # Calculate the number of days
    num_days = (end_date - start_date).days + 1

    # Day counter for identifying every 5th day as corporate
    day_counter = 0

    # Iterate over the number of days and calculate sales
    for day in range(num_days):
        current_date = start_date + timedelta(days=day)
        vehicles_sold_today = vehicles_sold_per_day(day_counter % 5 + 1)
        
        # Only count every 5th day as corporate
        if (day_counter % 5) == 4:
            # Add to total corporate sales
            total_corporate_sales += vehicles_sold_today
            
            # Append the sales information for the current day
            corporate_sales_info.append({
                "Date": current_date.strftime("%Y-%m-%d"),
                "Vehicles Sold": vehicles_sold_today
            })
        
        # Increment day counter
        day_counter += 1

    return corporate_sales_info, total_corporate_sales

# Display corporate sales information from April 1st to September 30th
corporate_sales_info, total_corporate_sales = calculate_corporate_sales("2023-04-01", "2023-09-30")
corporate_sales_info, total_corporate_sales
```
